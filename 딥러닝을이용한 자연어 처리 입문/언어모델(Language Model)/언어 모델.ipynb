# 언어 모델(Language Model)

- 단어 시퀀스(문장)에 확률을 할당하는 모델을 말한다
    - e.x) 어떤 문장이 있을 때, 기계가 문장을 판단하는 기능이 있다면? 성능이 뛰어난것!!

- 종류
    - 통계에 기반한 전통적인 언어 모델(Statistical Languagel Model, SLM)
        - **참고** 인공 신경망으로 인해 사용 용도가 줄었음!!
    - 인공 신경망을 이용한 방법

## 1. 언어 모델
- 단어 시퀀스(문장)에 확률을 할당(assign)하는 일을 하는 모델
    - 가장 자연스러운 단어 시퀀스를 찾아내는 모델
    - 방법
        - 이전 단어들이 주어졌을 때 다음 단어를 예측하는 것
        - 주어진 양쪽의 단어들로부터 가운데 비어있는 단어를 예측하는 모델(BERT)
    - 언어 모델이 단어들의 조합이 얼마나 적절한지, 또는 해당 문장이 얼마나 적합한지를 알려줌
    

## 2. 단어 시퀀스의 확률 할당
- 확률을 통해 보다 적절한 문장을 판단한다
## 예시
### a. 기계 번역(Machine Translation):
- P (나는 버스를 탔다) > P(나는 버스를 태운다)
    - : 언어 모델은 두 문장을 비교하여 좌측의 문장의 확률이 더 높다고 판단

### b. 오타 교정(Spell Correction)
- 선생님이 교실로 부리나케 <br> P(달려갔다) > P(잘려갔다)
    - : 언어 모델은 두 문장을 비교해 좌측의 문장의 확률이 더 높다고 판단

### c. 음성 인식 (Speech Recognition)
- P(나는 메롱을 먹는다) < P(나는 메론을 먹는다)
    - : 언어 모델은 두 문장을 비교해 우측의 문장의 확률이 더 높다고 판단



## 3. 주어진 이전 단어들로부터 다음 단어 예측하기

### A. 단어 시퀀스의 확률
- 하나의 단어를 w, 단어 시퀀스를 W라고 하면, n개의 단어가 등장하는 단어 시퀀스 W의 확률은 


$$P(W) = P(w_1,w_2,w_3,w_4,w_5,...w_n)$$



### B. 다음 단어 등장 확률
- n-1개의 단어가 나열된 상태에서 n 번째 단어의 확률(조건부확률)
$$P(w_n|w_1, ..., w_{n-1})$$
- e.x) 다선번쨰 단어의 확률
$$P(w_5|w_1, w_2, w_3, w_4)$$
- 전체 단어 시퀀스 W의 확률으모든 단어가 예측되고 나서만 알 수 있다.
$$P(W) = P(w_1,w_2,w_3,w_4,w_5,...,w_n) = \prod_{k=1}^n P(w_i|w_1,\dots,w_{i_1})$$


## 4. 언어 모델의 간단한 직관
- 후보지의 단어들을 확률에 기반하여 예측하고, 가장 높은 확률의 단어를 반환한다.
