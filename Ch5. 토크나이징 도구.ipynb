# 토크나이징
자연어 처리를 위해서는 우선 텍스트에 대한 정보를 단위별로 나누는 것이 일반적이다.
ex) 영화 리뷰 내용 예측인 경우  한 문장을 단어 단위로 쪼개어 분석할 수 있다.

즉 예측해야할 입력 정보를 하나의 특정 기본 단위로 자르는 것을 토크나이징이라고 한다.



## 영어 토크나이징 라이브러리

NLTK
- 영어 텍스트 전처리 작업을 하는데 많이 쓰이는 라이브러리

```python
# 단어 단위로 문장 자르기
# 특수 문자의 경우 따로 구분이 된다.  -> '(', ')', ','
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize
sentence = "Natural language processing (NLP) is a subfield of computer science, information engineering, and artificial intelligence concerned with the interactions between computers and human (natural) languages, in particular how to program computers to precess and analyze large amounts of natural language data"
print(word_tokenize(sentence))
```

```python
# 문장 단위로 자르기
from nltk.tokenize import sent_tokenize
paragraph = "Natural language processing (NLP) is a subfield of computer science, information engineering, and artificial intelligence concerned with the interactions between computers and human (natural) languages, in particular how to program computers to process and analyze large amounts of natural language data. Challenges in natural language processing frequently invovlve speech recognition, natural language understanding, and natural language generation."
print(sent_tokenize(paragraph))
```

Spacy
NLTK와 같은 오픈소스 라이브러리이다. 
- 영어 포함 8개 국어에 대한 자연어 전처리 모듈을 제공하고, 빠른 속도로 전처리할 수 있다.
- 전처리를 한번에 할 수 있다.

```python
import spacy

# spacy.load('en') 을 통해 토크나이징할 객체를 생성하여 nlp 변수에 할당한다.
nlp = spacy.load('en')
sentence = "Natural language processing (NLP) is a subfield of computer science, information engineering, and artificial intelligence concerned with the interactions between computers and human (natural) languages, in particular how to program computers to precess and analyze large amounts of natural language data"
doc=nlp(sentence)
```

```python
word_tokenized_sentence = [token.text for token in doc]
sentence_tokenized_list = [sent.text for sent in doc.sents]
print(word_tokenized_sentence)
print(sentence_tokenized_list)
```
